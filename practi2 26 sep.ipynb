```
import pandas as pd
from matplotlib import pyplot as plt
Data = {'Country': ['USA','Canada', 'Germany','UK','France'],
        'GDP_Per_Capita': [45000, 42000,52000,49000,47000],
        'Income_Per_Capita': [4000,5000,7000,55000,60000]
        }

df = pd.DataFrame(Data)
df.plot(x = 'Country', y=['GDP_Per_Capita', 'Income_Per_Capita'], kind = 'bar')
```

```
import numpy as np
from matplotlib import pyplot as plt
plt.figure(figsize=[15,10])
totalDeaths = [112596,937312,459971,279136,490597,749949]
totalRecovery=[1125966,6637312,65971,2716636,4059766, 44966]
activeCases=[113995,347973,239990,293600,347300]
country=['USA','Brazil', 'Russia','India','Italy']
x=np.arange(len(totalDeaths))
plt.bar(x,totalDeaths,color ='Black',width=0.25)
plt.bar(x + 0.25,totalRecovery,color ='g',width=0.25)
plt.bar(x + 0.25, activeCases,color ='b',width=0.25)

plt.legend(['totalDeaths','totalRecovery','activeCases'])
plt.xticks([i + 0.25 for i in  range(5)],country)
plt.title("bar plot")
plt.xlabel('Countrie')
plt.ylabel('Cases')
plt.savefig('5barplot.png')
plt.show()
```

```
from pandas import DataFrame
import matplotlib.pyplot as plt
Data={'Tasks':[300,500,700],
      'Task Type':['Tasks pending','Tasks ongoing','Tasks completed']
      }

df=DataFrame(Data)
df.set_index('Task Type',inplace=True)

df.plot.pie(y='Tasks',figsize=(10,10),autopct='1%.1f%%',startangle=90)
```

```
import numpy as np
import matplotlib.pyplot as plt

labels=['civil','electrical','mechanical','chemical']
sizes=[15,30,45,10]

explode=(0.1,0.1,0.1,0.4)
fig,ax=plt.subplots()
ax.pie(sizes,
       explode=explode,
       labels=labels,
       autopct='$1.1f%%',
       shadow=True,
       startangle=90)
ax.axis('equal')
ax.set_title('engineering Diciplines')
plt.show()
```
